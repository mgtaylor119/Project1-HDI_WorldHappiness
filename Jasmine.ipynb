```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress

# Load in CSV
HDI_Happiness_2021 = pd.read_csv("Resources/HDI_Happiness_2021.csv")

# Drop rows with null values (NaN) from the DataFrame
HDI_Happiness_2021 = HDI_Happiness_2021.dropna()
HDI_Happiness_2021
```

Country's Happiness Score vs. Human Development Index Scatter Plot 

```python
# Scatter plot of Happiness Index vs Human Development
x_values = HDI_Happiness_2021["Happiness Index"]
y_values = HDI_Happiness_2021["Human Development Index (2021)"]

plt.xlabel("Happiness Index")
plt.ylabel("Human Development Index")
plt.title("Happiness Index vs Human Development Index(2021)")
plt.scatter(x_values, y_values, marker="o", color="blue")
plt.show()
```
