```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress

# Load in CSV
HDI_Happiness_2021 = pd.read_csv("Resources/HDI_Happiness_2021.csv")

# Drop rows with null values (NaN) from the DataFrame
HDI_Happiness_2021 = HDI_Happiness_2021.dropna()
HDI_Happiness_2021
```

## Compute Linear Regression for Each Relationship

```python
# Define a function to create Linear Regression plots
def plot_linear_regression(x_values, y_values, title, text_coordinates):
    (slope, intercept, rvalue, pvalue, stderr) = linregress (x_values,y_values)
    regress_values = x_values * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + " x + " + str(round(intercept,2))
    
    # Plot the scatter plot of the data
    plt.scatter(x_values, y_values)

    # Plot the regression line
    plt.plot(x_values, regress_values, "r-")

    # Annotate the equation on the plot at the given text coordinates
    plt.annotate(line_eq, text_coordinates, fontsize=12, color="red")

    # Print the correlation coefficient (r-value)
    print(f"The r-value is: {rvalue}")

    # Show the plot
    plt.show()
```

### Life Expectancy vs. Happiness Rank Linear Regression Plot

```python
# Linear regression on Life Expectancy vs. Happiness Rank 
x_values = HDI_Happiness_2021["Rank"]
y_values = HDI_Happiness_2021["Life Expectancy at Birth (2021)"]

plt.xlabel("Happiness Rank")
plt.ylabel("Life Expectancy at Birth (years)")
plt.title("Life Expectancy vs. Happiness Rank (2021)")
plot_linear_regression(x_values, y_values, 'Linear Regression', (10,60))
```

**Discussion about the linear relationship:** The linear regression line indicates a moderate, negative correlation between Life Expectancy and Happiness. The negative slope of -0.14 suggests that as the Happiness Rank increases (indicating lower happiness levels), there is a tendency for the average life expectancy to decrease. In contrast, as the Happiness Rank decreases (indicating higher happiness levels), there is a tendency for the average life expectancy to increase.

### Education vs. Happiness Rank Linear Regression Plot

```python
# Linear regression on Education vs. Happiness Rank
x_values = HDI_Happiness_2021["Rank"]
y_values = HDI_Happiness_2021["Mean Years of Schooling (2021)"]

plt.xlabel("Happiness Rank")
plt.ylabel("Years of Schooling (avg)")
plt.title("Education vs. Happiness Rank (2021)")
plot_linear_regression(x_values, y_values, 'Linear Regression', (10,4))
```

**Discussion about the linear relationship:** The linear regression line indicates a moderate, negative correlation between Education and Happiness. The negative slope of -0.05 suggests that as the Happiness Rank increases (indicating lower happiness levels), the Average Years of Schooling tend to decrease. In contrast, as the Happiness Rank decreases (indicating higher happiness levels), the Average Years of Schooling tend to increase. In general, countries with lower happiness levels tend to have lower average years of schooling, while countries with higher happiness levels tend to have higher average years of schooling.

### Income vs. Happiness Rank Linear Regression Plot

```python
# Linear regression on Income vs. Happiness Rank
x_values = HDI_Happiness_2021["Rank"]
y_values = HDI_Happiness_2021["Gross National Income Per Capita (2021)"]

plt.xlabel("Happiness Rank")
plt.ylabel("Income Per Capita")
plt.title("Income vs. Happiness Rank (2021)")
plot_linear_regression(x_values, y_values, 'Linear Regression', (80,50000))
```

**Discussion about the linear relationship:** The linear regression line indicates a moderate, negative correlation between Income and Happiness. The negative slope of -366.53 suggests that as the Happiness Rank increases (indicating lower happiness levels), the Gross National Income Per Capita tends to decrease. In contrast, as the Happiness Rank decreases (indicating higher happiness levels), the Gross National Income Per Capita tends to increase. In general, countries with lower happiness levels tend to have lower income per capita, while countries with higher happiness levels tend to have higher income per capita.
