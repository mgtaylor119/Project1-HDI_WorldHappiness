```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress

# Load in CSV
HDI_Happiness_2021 = pd.read_csv("Resources/HDI_Happiness_2021.csv")

# Drop rows with null values (NaN) from the DataFrame
HDI_Happiness_2021 = HDI_Happiness_2021.dropna()
HDI_Happiness_2021
```

## Compute Linear Regression for Each Relationship

```python
# Define a function to create Linear Regression plots
def plot_linear_regression(x_values, y_values, title, text_coordinates):
    (slope, intercept, rvalue, pvalue, stderr) = linregress (x_values,y_values)
    regress_values = x_values * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + " x + " + str(round(intercept,2))
    
    # Plot the scatter plot of the data
    plt.scatter(x_values, y_values)

    # Plot the regression line
    plt.plot(x_values, regress_values, "r-")

    # Annotate the equation on the plot at the given text coordinates
    plt.annotate(line_eq, text_coordinates, fontsize=12, color="red")

    # Print the correlation coefficient (r-value)
    print(f"The r-value is: {rvalue}")

    # Show the plot
    plt.show()
```

### Life Expectancy vs. Happiness Index Linear Regression Plot

```python
# Linear regression on Life Expectancy vs. Happiness Index 
x_values = HDI_Happiness_2021["Happiness Index"]
y_values = HDI_Happiness_2021["Life Expectancy at Birth (2021)"]

plt.xlabel("Happiness Index")
plt.ylabel("Life Expectancy at Birth (years)")
plt.title("Life Expectancy vs. Happiness Index (2021)")
plot_linear_regression(x_values, y_values, 'Linear Regression', (5.5,55))
```

**Discussion about the linear relationship:** The linear regression line indicates a positive correlation between Life Expectancy and Happiness. The positive slope of 5.57 suggests that that, on average, for each one-unit increase in the Happiness Index, there is an expected increase of 5.57 years in Life Expectancy at Birth. In contrast, as the Happiness Index decreases (indicating lower happiness levels), the average life expectancy tends to decrease. The relationship between Life Expectancy and Happiness is statistically significant as indicated by the r-value of 0.7658840088647312, showing a relatively strong correlation between the two variables.

### Education vs. Happiness Index Linear Regression Plot

```python
# Linear regression on Education vs. Happiness Index
x_values = HDI_Happiness_2021["Happiness Index"]
y_values = HDI_Happiness_2021["Mean Years of Schooling (2021)"]

plt.xlabel("Happiness Index")
plt.ylabel("Years of Schooling (avg)")
plt.title("Education vs. Happiness Index (2021)")
plot_linear_regression(x_values, y_values, 'Linear Regression', (5.5,3))
```

**Discussion about the linear relationship:** The linear regression line indicates a moderate, positive correlation between Education and Happiness. The positive slope of 2.11 suggests that as the Happiness Index increases (higher happiness levels), there is a moderate tendency for the average years of schooling to increase. In general, countries with higher Happiness Index scores tend to have higher average years of schooling, and countries with lower Happiness Index scores tend to have lower average years of schooling. The relationship between Education and Happiness is statistically significant as indicated by the r-value of 0.6985499956715034, showing a moderate correlation between the two variables.

### Income vs. Happiness Index Linear Regression Plot

```python
# Linear regression on Income vs. Happiness Index
x_values = HDI_Happiness_2021["Happiness Index"]
y_values = HDI_Happiness_2021["Gross National Income Per Capita (2021)"]

plt.xlabel("Happiness Index")
plt.ylabel("Income Per Capita")
plt.title("Income vs. Happiness Index (2021)")
plot_linear_regression(x_values, y_values, 'Linear Regression', (4.5,-15000))
```

**Discussion about the linear relationship:** The positive slope of the regression line (14451.34) indicates a direct relationship between Income and Happiness. For each one-unit increase in the Happiness Index, the Income Per Capita is expected to increase by 14,451.34 (approximately $14,451.34). Conversely, has the Happiness Index decreases, the Gross National Income Per Capita tends to decrease. The relationship between Income and Happiness is statistically significant as indicated by the r-value of 0.7745149054243162, showing a relatively strong correlation between the two variables.
