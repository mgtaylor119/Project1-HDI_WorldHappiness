```python
locations_data = pd.read_csv(r"C:\Users\nicho\OneDrive\Documents\GitHub\Project1-HDI_WorldHappiness\resources\world_country_and_usa_states_latitude_and_longitude_values.csv")
```

```python
map_set = pd.merge(mapping_data, locations_data, how="right", on = ["Country", "Country"])
```

```python
map_set2 = map_set.drop(columns = ["country_code", "usa_state_code", "usa_state_latitude", "usa_state_longitude", "usa_state"]).copy()
```

```python
map_set2
```

```python
null_drop = []

for i in range(len(map_set2)):
    if map_set2["Rank"][i] == "NaN":
        map_set2.append(i)
    
mapping_data = map_set2.drop(null_drop)
mapping_data.dropna(inplace = True)
```

```python
mapping_data
```

```python
mapping_data.sort_values(["Rank","Country"],axis=0, ascending=True,inplace=True,na_position='first')
```

```python


map_plot_2 = mapping_data.hvplot.points(
    "longitude",
    "latitude",
    #geo = True,
    tiles = "ESRI",
    title = "World Happiness by Country Rank",
    size = "Rank",
    #scale = 2,
    color = "Rank",
    clabel = "Happy....................Moderate..................Unhappy",
    hover = False
)
colormaps = hv.plotting.list_cmaps()
spacing = np.linspace(0, 1, 64)[np.newaxis]
opt_kwargs = dict(aspect=6, xaxis=None, yaxis=None, sublabel_format='')

def filter_cmaps(category):
    return hv.plotting.util.list_cmaps(records=True,category=category,reverse=False)

def cmap_examples(category,cols=4):
    cms = filter_cmaps('bmw')
    n = len(cms)*1.0
    c=ceil(n/cols) if n>cols else cols
    bars = [hv.Image(spacing, ydensity=1, label="{0} ({1})".format(r.name,r.provider))\
            .opts(cmap=process_cmap(r.name,provider=r.provider), **opt_kwargs)
           for r in cms]
    return hv.Layout(bars).opts(vspace=0.1, hspace=0.1, transpose=(n>cols)).cols(c)

hv.extension('bokeh')

defaults = dict(width=400, height=400)
hv.opts.defaults(
    opts.EdgePaths(**defaults), opts.Graph(**defaults), opts.Nodes(**defaults))

map_plot_2
```
